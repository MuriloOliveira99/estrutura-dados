```python
from faker import Faker
import pandas as pd
```

## Gerando dados fake

```python
faker = Faker('pt-BR')

dados = [
    {
        "nome": faker.first_name(),
        "sobrenome": faker.last_name(),
        "endereco": faker.city(),
        "estado": faker.state()
    }
    for _ in range(100)
]

df = pd.DataFrame(dados)
```

```python
def gerar_dados(n):
    '''
    faker.unique.state()
    '''
    dados = {
        "nome": [faker.name() for _ in range(n)],
        "estado": [faker.state() for _ in range(n)],
        "bairro": [faker.city() for _ in range(n)]
    }

    return pd.DataFrame(dados)
```

```python
df = gerar_dados(10)
df.head(20)
```

```python
df = pd.concat([df, pd.DataFrame(gerar_dados(30))], ignore_index=True)
```
